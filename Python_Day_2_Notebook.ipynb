# While loop

```python
password = 123
user_entry = int(input("Please enter your password:"))

while password != user_entry:
  print("Password is incorrect. Try again.")
  user_entry = int(input("Please enter your password:"))
  if password == user_entry:
    print("Access is granted.")
    break
```
